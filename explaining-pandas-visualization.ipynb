```python
list_dfs = pd.read_html("https://en.wikipedia.org/wiki/Vilnius")
```

```python
df = list_dfs[5]
```

```python
df = df.rename(columns={"Area (km2)": "Area", "Population[237]": "Population", "Density (per km2)": "Density"})
```

```python
df
```

```python
df["Density2"] = df["Population"] / df["Area"]
df
```

```python
total_area = df["Area"].sum()
total_population = df["Population"].sum()
total_density = total_population / total_area
print("Total area in Vilnius:", total_area)
print("Total population in Vilnius:", total_population)
print("Total population density in Vilnius:", total_density)
```

```python
biggest_area = df.iloc[df['Area'].idxmax()]
smallest_area = df.iloc[df['Area'].idxmin()]
most_populous_area = df.iloc[df['Population'].idxmax()]
least_populous_area = df.iloc[df['Population'].idxmin()]
```

```python
biggest_area["Eldership"]
```

```python
print(f"What is the biggest area in Vilnius? {biggest_area['Eldership']} Size: {biggest_area['Area']}")
print(f"What is the smallest area in Vilnius? {smallest_area['Eldership']} Size: {smallest_area['Area']}")
print(f"What is the most populous area in Vilnius? {most_populous_area['Eldership']} Size: {most_populous_area['Population']}")
print(f"What is the least populous area in Vilnius? {least_populous_area['Eldership']} Size: {least_populous_area['Population']}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# From 0 to 100, generate 100 equally-spaced points
xs = np.linspace(0, 100, num=1000)
ys = [np.sin(x) for x in xs]
plt.figure(figsize=(15, 8))
plt.plot(xs, ys, marker="o", label="f(x) = x^2")
plt.grid()
plt.legend()
plt.show()
```
